### 8.2.1 读取数据集

```python
lines = d2l.read_time_machine()
print(f'# 文本总行数: {len(lines)}')
print(lines[0])
print(lines[10])
```

### 8.2.2 词元化

```python
tokens = d2l.tokenize(lines)
for i in range(11):
    print(tokens[i])
```

### 8.2.3 词表

```python
vocab = d2l.Vocab(tokens)
print(list(vocab.token_to_idx.items())[:10])
```

```python
for i in [0, 10]:
    print('文本: ', tokens[i])
    print('索引: ', vocab[tokens[i]])
```
